````python
# One of the most thing let us using numpy is the optimized memory and time

import time
import sys
import random

array1 = np.random.rand(1000000)
array2 = np.random.rand(1000000)

start_time = time.time()
result = array1 + array2
end_time = time.time()

# Calculate memory usage for NumPy version
memory_usage_numpy = sys.getsizeof(result)

# Print execution time and memory usage for NumPy version
print("NumPy Version:")
print("Execution Time:", end_time - start_time, "seconds")
print("Memory Usage:", memory_usage_numpy, "bytes")

# Create large Python lists
list1 = [random.random() for _ in range(1000000)]
list2 = [random.random() for _ in range(1000000)]

# Measure execution time for Python loop version
start_time = time.time()
result = [list1[i] + list2[i] for i in range(len(list1))]
end_time = time.time()

# Calculate memory usage for Python loop version
memory_usage_python = sys.getsizeof(result)

# Print execution time and memory usage for Python loop version
print("\nPython Loop Version:")
print("Execution Time:", end_time - start_time, "seconds")
print("Memory Usage:", memory_usage_python, "bytes")
````

## Scrapping data from a website

````python
# BeautifulSoup library (often imported as "bs4"). BeautifulSoup is a popular Python library used for web scraping and parsing HTML and XML documents.
!pip install bs4
````

````python
from bs4 import BeautifulSoup
import requests
````

````python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
````

````python
page
````

````python
page.text
````

````python
soup.find_all('p')
````

````python
#find and extract the first <table> element from the HTML content stored in the soup object.
soup.find('table')
````

````python
len(soup.find_all('table'))
````

````python
soup.find_all('table')[3]
````

````python
#get the element in the HTML content stored in the soup object that has both the class "wikitable" and "sortable.
soup.find('sup', class_ = 'wikitable sortable')
````

````python
table = soup.find_all('table')[1]
````

````python
print(table)
````

````python
world_titles = table.find_all('th')
````

````python
world_titles
````

````python
world_table_titles = [title.text.strip() for title in world_titles]

print(world_table_titles)
````

````python
import pandas as pd
````

````python
df=pd.DataFrame(columns = world_table_titles)

df
````

````python
column_data = table.find_all('tr')
````

````python
column_data
````

````python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
````

````python
df
````

````python
df.to_csv("h.csv")
````
